```python
import numpy as np
import pandas as ps
```

```python
df = ps.read_csv('fashion-mnist_train.csv')
df
```

```python
y = df['label'].as_matrix()
y = np.reshape(y,(len(y),1))

df = df.drop(['label'], axis = 1)
x = df.as_matrix() / 255
```

```python
from keras.utils import np_utils
y = np_utils.to_categorical(y, num_classes=10)
print(y)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
def learn():
    iterations = 10000
    learning_rate = 0.0001
    w1 = np.random.randn(784,15)
    w2 = np.random.randn(15,10)
    for i in range(iterations):
        # Forward propagation
        z1 = np.dot(x, w1)
        a1 = sigmoid(z1)
        z2 = np.dot(a1, w2)
        a2 = sigmoid(z2)
        
        # Loss (crossentropy se potriveste mai bine)
        
        error = np.sum((a2 - y) * (a2 - y))
        
        # Back propagation
        
        derror_da2 = a2 - y
        da2_dz2 = d_sigmoid(z2)
        dz2_da1 = w2
        dz2_dw2 = a1
        da1_dz1 = d_sigmoid(z1)
        dz1_dw1 = x
        
        derror_dw2 = np.dot(dz2_dw2.T, derror_da2 * da2_dz2)
        derror_dw1 = np.dot(dz1_dw1.T, np.dot(derror_da2 * da2_dz2, dz2_da1.T) * da1_dz1)
        
        w2 -= learning_rate * derror_dw2
        w1 -= learning_rate * derror_dw1
        
        if i % 100 == 1:
            print(error)
    return w1,w2
```

```python
w1, w2 = learn()
```
